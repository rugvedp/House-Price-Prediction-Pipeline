```python
from sklearn.impute import SimpleImputer 
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OrdinalEncoder 

from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
```

```python
data = pd.read_csv('../src/dataset/Housing.csv')
```

```python
X=data.drop(labels=["guestroom","price","hotwaterheating","airconditioning","prefarea"],axis=1)
y=data["price"]
```

```python
X.head()
```

```python
y.head()
```

```python
cat_cols=X.select_dtypes(include="object").columns

num_cols=X.select_dtypes(exclude="object").columns
```

```python
mainroad_cat = ['yes','no']
basement_cat = ['yes','no']
furnished_cat = ["unfurnished","semi-furnished","furnished"]
```

```python
num_pipeline=Pipeline(
    steps=[
        ("imputer",SimpleImputer()),
        ("scaler",StandardScaler())
    ]
)
```

```python
cat_pipeline=Pipeline(

    steps=[
        ("imputer",SimpleImputer(strategy="most_frequent")),
        ("ordinalencoder",OrdinalEncoder(categories=[mainroad_cat,basement_cat,furnished_cat]))

    ]
)
```

```python
cat_pipeline
```

```python
preprocessor=ColumnTransformer(

    [
        ("num_pipeline",num_pipeline,num_cols),
        ("cat_pipeline",cat_pipeline,cat_cols)
    ]
)
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30, random_state=42)
```

```python
X_train=pd.DataFrame(preprocessor.fit_transform(X_train),columns=preprocessor.get_feature_names_out())
X_test=pd.DataFrame(preprocessor.transform(X_test),columns=preprocessor.get_feature_names_out())
```

```python
from xgboost import XGBRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
models={

    'LinearRegression':LinearRegression(),
    'Randomforest':RandomForestRegressor(),
    'xgboost':XGBRegressor()

}
```

```python
trained_model_list=[]
model_list=[]
r2_list=[]
```

```python
def evaluate_model(true,pred):
    r2=r2_score(true,pred)
    mae=mean_absolute_error(true,pred)
    mse=mean_squared_error(true,pred)
    return mae, mse,r2
```

```python
for i in range(len(list(models))):
    model=list(models.values())[i]
    
    model.fit(X_train,y_train)

    y_pred=model.predict(X_test)

    MAE,MSE,R2=evaluate_model(y_test,y_pred)


    print("model training performance",model)
    print("MSE:", MSE)
    print("MAE:",MAE)
    print("R2 SCORE:",R2)

    r2_list.append(R2)

    print("="*40)
    print("\n")
```
